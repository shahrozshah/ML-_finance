```python
import pandas as pd


sid_last_digit = 2


n = sid_last_digit
print("n:", n)
```

```python
data = {
    'relationship': ['Husband', 'Not-in-family'],
    'hours-per-week': [40, 80  ]
}
df = pd.DataFrame(data)


grouped_df = df.groupby(['relationship', 'hours-per-week']).size().reset_index(name='count')
print("Grouped by 'relationship' and 'hours-per-week':")
print(grouped_df)

```

```python

grouped_reduced_df = df.groupby(['relationship', 'reduced-hours-per-week']).size().reset_index(name='count')
print("Grouped by 'relationship' and reduced 'hours-per-week':")
print(grouped_reduced_df)
```

```python
df['reduced-hours-per-week'] = df['hours-per-week'] - n
print("DataFrame with reduced 'hours-per-week':")
print(df)
```

```python
grouped_reduced_df = df.groupby(['relationship', 'reduced-hours-per-week']).size().reset_index(name='count')
print("Grouped by 'relationship' and reduced 'hours-per-week':")
print(grouped_reduced_df)
```
